```
import pandas as pd

def recommend_by_genre(df, preferred_genre, top_n=5):
    """Recommends top N movies based on the user's preferred genre, sorted by rating."""

    # Normalize genre column for consistent filtering
    df_exploded = df.copy()
    df_exploded['Genere'] = df_exploded['Genere'].apply(lambda x: [g.strip().lower() for g in x])  # Clean genres
    df_exploded = df_exploded.explode('Genere')  # Separate multiple genres into rows

    # Normalize user input (remove spaces, make lowercase)
    preferred_genre = preferred_genre.strip().lower()

    # Filter movies based on preferred genre
    genre_movies = df_exploded[df_exploded['Genere'] == preferred_genre]

    # If no movies are found for the genre
    if genre_movies.empty:
        return "No movies found for the specified genre."

    # Sort movies by Rating in descending order and return top N
    recommended_movies = genre_movies.sort_values(by='Rating', ascending=False).head(top_n)

    return recommended_movies[['Movie Names', 'Rating', 'Genre']]

# Load dataset (Replace with actual file path)
os.chdir("/content/drive/My Drive/Colab Notebooks")
os.getcwd()
df = pd.read_csv("Indian_movies.csv")

# Convert 'Rating' to numeric (handles errors by setting invalid values to NaN)
df['Rating'] = pd.to_numeric(df['Rating'], errors='coerce')

# Drop rows where 'Rating' is NaN
df.dropna(subset=['Rating'], inplace=True)

# Convert 'Genre' column from a comma-separated string to a list (ONLY if necessary)
if isinstance(df['Genere'].iloc[0], str):  # Check if the first value is a string
    df['Genere'] = df['Genere'].str.split(', ')  # Convert to list

# Get user input for preferred genre
preferred_genre = input("Enter your preferred genre: ")

# Get recommendations
recommendations = recommend_by_genre(df, preferred_genre)

# Print results
print("\nRecommended Movies:")
print(recommendations)
```

```
import pandas as pd

def recommend_by_genre(df, preferred_genre, top_n=5):
    """Recommends top N movies based on the user's preferred genre, sorted by rating."""

    # Ensure 'Genre' column is formatted correctly
    df['Genre'] = df['Genre'].apply(lambda x: [g.strip() for g in x if g.strip().lower() != "back to top"])  # Remove "Back to top"

    # Explode the 'Genre' column (separates multiple genres into rows)
    df_exploded = df.explode('Genre')

    # Normalize user input (remove spaces, make lowercase)
    preferred_genre = preferred_genre.strip()

    # Filter movies based on the preferred genre
    genre_movies = df_exploded[df_exploded['Genre'] == preferred_genre]

    # If no movies are found for the genre
    if genre_movies.empty:
        return "No movies found for the specified genre."

    # Sort movies by Rating in descending order and return top N
    recommended_movies = genre_movies.sort_values(by='Rating', ascending=False).head(top_n)

    return recommended_movies[['Movie Names', 'Rating', 'Genre']]

# Load dataset from JSON (replace with your actual file path)
os.chdir("/content/drive/My Drive/Colab Notebooks")
os.getcwd()
df = pd.read_csv("Indian_movies.csv")


# Rename incorrect column
df.rename(columns={'Genere': 'Genre'}, inplace=True)

# Convert 'Genre' column from a comma-separated string to a list
df['Genre'] = df['Genre'].str.split(',')

# Get user input for preferred genre
preferred_genre = input("Enter your preferred genre: ")

# Get recommendations
recommendations = recommend_by_genre(df, preferred_genre)

# Print results
print("\nRecommended Movies:")
print(recommendations)
```
